### Data Preprocessing


### PREPROCESSING: Data preprocessing is a data mining technique which is used to transform the raw data in a useful and efficient format

```python
import pandas as pd
df = pd.read_excel('Iris.xlsx', usecols="B,C,D,E")
print(df)
```

```python
import matplotlib.pyplot as plt
df.plot(kind = 'line')
```

```python
#Normalizing data
df_max_scaled = df.copy()
for column in df_max_scaled.columns:
	df_max_scaled[column] = df_max_scaled[column] / df_max_scaled[column].abs().max()	
print(df_max_scaled)
```

```python
df_max_scaled.plot(kind = 'bar')
```

```python
dfminmaxscaled = df.copy()
for column in dfminmaxscaled.columns:
    dfminmaxscaled[column] = (dfminmaxscaled[column] - dfminmaxscaled[column].min()) / (dfminmaxscaled[column].max() - dfminmaxscaled[column].min())    
print(dfminmaxscaled)
```

```python
dfminmaxscaled.plot(kind = 'bar')
```

```python
df_z_scaled = df.copy()
for column in df_z_scaled.columns:
	df_z_scaled[column] = (df_z_scaled[column] - df_z_scaled[column].mean()) / df_z_scaled[column].std()	

print(df_z_scaled)
```

```python
df_z_scaled.plot(kind='bar')
```

### Binning Method: 
### This method works on sorted data in order to smooth it.
### The whole data is divided into segments of equal size and then various methods are performed to complete the task.
### One can replace all data in a segment by its mean or boundary values can be used to complete the task. 

```python
#Smoothing data by Binning
import numpy as np
import math

a = np.zeros(150)

for i in range(150):
	a[i] = df["SepalLengthCm"][i]

b = np.zeros(150)

for i in range (150):
	b[i]=a[i]

b=np.sort(b)


bin1=np.zeros((30,5))
bin2=np.zeros((30,5))
bin3=np.zeros((30,5))


for i in range (0,150,5):
	k=int(i/5)
	mean=(b[i] + b[i+1] + b[i+2] + b[i+3] + b[i+4])/5
	for j in range(5):
		bin1[k,j]=mean
print("Bin Mean: \n",bin1)
	

for i in range (0,150,5):
	k=int(i/5)
	for j in range (5):
		if (b[i+j]-b[i]) < (b[i+4]-b[i+j]):
			bin2[k,j]=b[i]
		else:
			bin2[k,j]=b[i+4]	
print("Bin Boundaries: \n",bin2)


for i in range (0,150,5):
	k=int(i/5)
	for j in range (5):
		bin3[k,j]=b[i+2]
print("Bin Median: \n",bin3)
```

```python
#Binarizing data
from sklearn import preprocessing
SepalLengthCm = df.iloc[:, 0].values
SepalWidthCm = df.iloc[:, 1].values
PetalLengthCm = df.iloc[:, 2].values
PetalWidthCm = df.iloc[:, 3].values

from sklearn.preprocessing import Binarizer

x = SepalLengthCm
x = x.reshape(1, -1)
y = SepalWidthCm
y = y.reshape(1, -1)
z = PetalLengthCm
z = z.reshape(1,-1)
w = PetalWidthCm
w = w.reshape(1,-1)

binarizer_1 = Binarizer(threshold=5,copy=False)
binarizer_2 = Binarizer(threshold=3,copy=False)
binarizer_3 = Binarizer(threshold=2,copy=False)
binarizer_4 = Binarizer(threshold=1,copy=False)
 

print (binarizer_1.fit_transform(x))
print (binarizer_2.fit_transform(y))
print (binarizer_3.fit_transform(z))
print (binarizer_4.fit_transform(w))
```

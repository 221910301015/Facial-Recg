```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv(r"C:\Users\adith\OneDrive\Desktop\Adithya\My Projects\Linear Reg\Advertising.csv")
df
```

```python
df.corr()  # Correlation to check more than one feature dependent to obtain optimal solution
```

```python
x=df[['TV','Radio']].values
y=df['Sales'].values.reshape(-1,1)
x
```

```python
lin=LinearRegression()
lin.fit(x,y)
```

```python
Pred_y=lin.predict(x)
Pred_y
```

```python
from sklearn import metrics
print(metrics.mean_absolute_error(y,Pred_y))
print(metrics.mean_squared_error(y,Pred_y))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv(r"C:\Users\adith\OneDrive\Desktop\Adithya\My Projects\Linear Reg\Advertising.csv")
df
```

```python
x=df['TV'].values.reshape(-1,1)
y=df['Sales'].values.reshape(-1,1)
y
```

```python
lin=LinearRegression()
lin.fit(x,y)
Pred_y=lin.predict(x)
Pred_y
```

```python
np.concatenate((y,Pred_y),axis=1)
```

#### MAE- This measures absolute averege distance between real data and predicted data 
#### MSE-This measures squared average distance between real data and predicted data

```python
((y-Pred_y)**2).mean()  #MSE
```

```python
np.abs(y-Pred_y).mean() #MAE
```

```python
from sklearn import metrics
print(metrics.mean_absolute_error(y,Pred_y))
print(metrics.mean_squared_error(y,Pred_y))
```
